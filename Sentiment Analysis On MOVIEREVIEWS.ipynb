```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
labels = [1 if 'pos' in file_id else 0 for file_id in movie_reviews.fileids()]
```

```python
X_train, X_test, y_train, y_test = train_test_split(preprocessed_reviews, labels, test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create a TF-IDF representation of the text data
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

```python

# Train a Naive Bayes classifier using TF-IDF features
classifier_tfidf = MultinomialNB()
classifier_tfidf.fit(X_train_tfidf, y_train)
```

```python
y_pred_tfidf = classifier_tfidf.predict(X_test_tfidf)
```

```python
accuracy_tfidf = accuracy_score(y_test, y_pred_tfidf)
print(f"Accuracy (TF-IDF): {accuracy_tfidf:.2%}")
```

```python
print("Classification Report (TF-IDF):")
print(classification_report(y_test, y_pred_tfidf))
```

```python
from sklearn.model_selection import GridSearchCV

# Define the hyperparameter grid for MultinomialNB
param_grid = {
    'alpha': [0.1, 0.5, 1.0, 2.0],  # Smoothing parameter
    'fit_prior': [True, False]  # Whether to learn class prior probabilities
}

# Create a TF-IDF representation of the text data
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

```python
# Initialize the Naive Bayes classifier
naive_bayes = MultinomialNB()

# Create a GridSearchCV object
grid_search = GridSearchCV(naive_bayes, param_grid, cv=5, scoring='accuracy', n_jobs=-1)

# Perform grid search on the training data
grid_search.fit(X_train_tfidf, y_train)
```

```python
# Perform grid search on the training data
grid_search.fit(X_train_tfidf, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_
```

```python
best_naive_bayes = MultinomialNB(**best_params)
best_naive_bayes.fit(X_train_tfidf, y_train)
```

```python
y_pred_tuned = best_naive_bayes.predict(X_test_tfidf)
```

```python
accuracy_tuned = accuracy_score(y_test, y_pred_tuned)
print(f"Accuracy (Tuned): {accuracy_tuned:.2%}")

# Display the best hyperparameters
print("Best Hyperparameters:", best_params)
```
